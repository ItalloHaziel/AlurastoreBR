# LOJA 1


```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados da loja 1
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
loja1 = pd.read_csv(url)

# Converter para numérico (se necessário)
loja1['Preço'] = pd.to_numeric(loja1['Preço'], errors='coerce')
loja1['Frete'] = pd.to_numeric(loja1['Frete'], errors='coerce')

# Criar coluna de Faturamento
loja1['Faturamento'] = loja1['Preço'] + loja1['Frete']

# Agrupar por categoria e somar o faturamento
faturamento_por_categoria = loja1.groupby('Categoria do Produto')['Faturamento'].sum().sort_values(ascending=False)

# Dividir faturamento por mil para melhor leitura
faturamento_por_categoria_mil = faturamento_por_categoria / 1000

# Plotar gráfico
plt.figure(figsize=(20, 8))
ax = faturamento_por_categoria_mil.plot(kind='bar', color='green')

# Adicionar valores dentro das barras
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}K',  # Exibir o valor com "K" para mil
                (p.get_x() + p.get_width() / 2., p.get_height()),  # Coordenadas
                xytext=(0, 10),  # Deslocamento do texto
                textcoords='offset points',  # Usar deslocamento em pontos
                ha='center', va='center', fontsize=10, color='black')

plt.title('Faturamento por Categoria de Produto')
plt.xlabel('Categoria')
plt.ylabel('Faturamento (milhões R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados da loja 1
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
loja1 = pd.read_csv(url1)

# Contar a quantidade de vezes que cada produto foi vendido
vendas_por_produto = loja1['Produto'].value_counts()

# Top 10 produtos mais vendidos
top_mais_vendidos = vendas_por_produto.sort_values(ascending=False).head(10)

# Top 10 produtos menos vendidos
top_menos_vendidos = vendas_por_produto.sort_values(ascending=True).head(10)

# Criar figura com dois subgráficos
fig, axes = plt.subplots(1, 2, figsize=(20, 6))

# Plotar gráfico dos mais vendidos
ax1 = top_mais_vendidos.sort_values().plot(kind='barh', color='green', ax=axes[0])

# Adicionar as quantidades dentro das barras para os mais vendidos
for i in ax1.patches:
    ax1.text(i.get_width() + -1.5, i.get_y() + i.get_height()/2,
             str(int(i.get_width())), va='center', ha='left', fontsize=10, color='black')

ax1.set_title('Top 10 Produtos Mais Vendidos - Loja 1')
ax1.set_xlabel('Quantidade de Vendas')
ax1.set_ylabel('Produto')
ax1.grid(axis='x', linestyle='--', alpha=0.7)

# Plotar gráfico dos menos vendidos
ax2 = top_menos_vendidos.sort_values(ascending=False).plot(kind='barh', color='salmon', ax=axes[1])

# Adicionar as quantidades dentro das barras para os menos vendidos
for i in ax2.patches:
    ax2.text(i.get_width() + -1, i.get_y() + i.get_height()/2,
             str(int(i.get_width())), va='center', ha='left', fontsize=10, color='black')

ax2.set_title('Top 10 Produtos Menos Vendidos - Loja 1')
ax2.set_xlabel('Quantidade de Vendas')
ax2.set_ylabel('Produto')
ax2.grid(axis='x', linestyle='--', alpha=0.7)

# Ajustar layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs dos arquivos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Leitura dos dados
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Adiciona o nome da loja
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Une todos os DataFrames
df = pd.concat([loja1, loja2, loja3, loja4])

# Converte a coluna de data
df['Data da Compra'] = pd.to_datetime(df['Data da Compra'])

# Cria a coluna de trimestre
df['Trimestre'] = df['Data da Compra'].dt.to_period('Q')

# Agrupa por loja e trimestre, somando os preços
df_grouped = df.groupby(['Loja', 'Trimestre'])['Preço'].sum().reset_index()

# Pivot para o formato de gráfico
pivot_df = df_grouped.pivot(index='Trimestre', columns='Loja', values='Preço')

# Plotagem
pivot_df.plot(kind='bar', figsize=(25, 6))
plt.title('Vendas por Trimestre por Loja')
plt.xlabel('Trimestre')
plt.ylabel('Vendas (R$)')
plt.xticks(rotation=45)
plt.legend(title='Loja')
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap

# URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# Leitura dos dados
df = pd.concat([
    pd.read_csv(url1),
    pd.read_csv(url2),
    pd.read_csv(url3),
    pd.read_csv(url4)
], ignore_index=True)

# Dicionário com coordenadas médias de cada estado do Brasil
coordenadas_estados = {
    'AC': [-9.02, -70.81],
    'AL': [-9.57, -36.77],
    'AM': [-3.07, -61.66],
    'AP': [1.41, -51.77],
    'BA': [-12.96, -38.51],
    'CE': [-5.20, -39.53],
    'DF': [-15.83, -47.86],
    'ES': [-19.19, -40.34],
    'GO': [-15.98, -49.86],
    'MA': [-5.42, -45.44],
    'MG': [-18.10, -44.38],
    'MS': [-20.51, -54.54],
    'MT': [-12.64, -55.42],
    'PA': [-3.79, -52.48],
    'PB': [-7.06, -35.55],
    'PE': [-8.38, -37.86],
    'PI': [-7.06, -42.82],
    'PR': [-24.89, -51.55],
    'RJ': [-22.84, -43.15],
    'RN': [-5.81, -36.59],
    'RO': [-11.22, -62.80],
    'RR': [1.89, -61.22],
    'RS': [-30.01, -51.22],
    'SC': [-27.33, -49.44],
    'SE': [-10.57, -37.45],
    'SP': [-23.55, -46.63],
    'TO': [-10.25, -48.25]
}

# Agora associamos cada ponto (lat/lon) a um estado com base em proximidade (simplificação)
# Vamos converter os pontos para o estado mais próximo (não precisa de API externa)
from scipy.spatial import distance

# Função para encontrar estado mais próximo para uma coordenada
def encontrar_estado(lat, lon):
    min_dist = float('inf')
    estado_mais_proximo = None
    for estado, coord in coordenadas_estados.items():
        dist = distance.euclidean((lat, lon), coord)
        if dist < min_dist:
            min_dist = dist
            estado_mais_proximo = estado
    return estado_mais_proximo

# Aplica função para cada linha
df['estado'] = df.apply(lambda row: encontrar_estado(row['lat'], row['lon']), axis=1)

# Conta número de vendas por estado
vendas_por_estado = df['estado'].value_counts().reset_index()
vendas_por_estado.columns = ['estado', 'quantidade_vendas']

# Junta com coordenadas
vendas_por_estado['lat'] = vendas_por_estado['estado'].map(lambda x: coordenadas_estados[x][0])
vendas_por_estado['lon'] = vendas_por_estado['estado'].map(lambda x: coordenadas_estados[x][1])

# Criar mapa
mapa = folium.Map(location=[-14.2350, -51.9253], zoom_start=4)

# Cria lista de pontos para o heatmap
heat_data = [[row['lat'], row['lon'], row['quantidade_vendas']] for _, row in vendas_por_estado.iterrows()]

# Adiciona o HeatMap
HeatMap(heat_data, radius=35, max_zoom=6).add_to(mapa)

# Exibe o mapa
mapa
```

Análise Comparativa das Lojas da Alura Store
Com o objetivo de identificar qual loja da rede Alura Store deve ser vendida para viabilizar um novo empreendimento, foi realizada uma análise abrangente de desempenho das quatro unidades da rede. Foram considerados os seguintes critérios:

Faturamento total

Avaliações dos clientes

Custo médio de frete

Categorias e produtos mais vendidos

Evolução trimestral das vendas (ver gráfico abaixo)

1. Faturamento Total
A métrica de faturamento é um dos principais indicadores financeiros. A seguir, os resultados consolidados por loja:

Loja 1: R$ 1.616.347,09

Loja 2: R$ 1.567.773,22

Loja 3: R$ 1.542.047,69

Loja 4: R$ 1.458.253,46

A Loja 1 apresentou o maior faturamento da rede, com desempenho superior em valor bruto de vendas. Já a Loja 4 obteve o pior resultado financeiro, com uma diferença de cerca de R$ 158 mil em relação à loja líder, o que demonstra menor capacidade de geração de receita.

2. Avaliação dos Clientes
As avaliações refletem a experiência do consumidor, impactando diretamente na fidelização e imagem da loja:

Loja 1: 3,98

Loja 2: 4,04

Loja 3: 4,05

Loja 4: 4,00

A Loja 3 possui a melhor média de avaliação, seguida pela Loja 2. A Loja 1, apesar do bom faturamento, tem a pior nota de satisfação, o que pode indicar fragilidades no serviço ou entrega. A Loja 4, novamente, aparece apenas com desempenho mediano, não se destacando positivamente em nenhum aspecto.

3. Frete Médio
O custo médio de frete impacta tanto a atratividade da loja quanto a margem de lucro:

Loja 1: R$ 34,69

Loja 2: R$ 33,62

Loja 3: R$ 33,07

Loja 4: R$ 31,28

A Loja 4 se destaca com o menor frete médio, o que, teoricamente, poderia ser um diferencial competitivo. Contudo, esse benefício não se refletiu em aumento de vendas ou avaliações, sugerindo que outros fatores têm maior peso para os consumidores.

4. Categorias e Produtos Vendidos
As lojas compartilham perfis semelhantes de produtos, com maior participação de itens eletrônicos, utilidades domésticas e acessórios. No entanto, a capacidade de conversão desse mix em faturamento é melhor na Loja 1, o que demonstra maior eficiência de venda por categoria.

5. Análise de Desempenho Trimestral (Gráfico)


O gráfico acima mostra as vendas trimestrais das quatro lojas ao longo de três anos. A Loja 1 mantém relativa estabilidade e picos em momentos importantes, enquanto a Loja 4 demonstra forte queda de desempenho ao longo do tempo, com tendência clara de redução de faturamento desde o início de 2022. A Loja 3 apresentou um pico pontual (Q4 de 2021), mas não sustentou o crescimento. Já a Loja 2 demonstra maior estabilidade e consistência ao longo dos períodos.

Conclusão e Recomendação Final
Com base em todas as análises apresentadas, a Loja 4 é a unidade com o pior desempenho geral:

Menor faturamento

Avaliação apenas mediana

Frete mais barato que não se converte em maior volume de vendas

Queda contínua no desempenho trimestral

Mesmo com alguma vantagem no frete e avaliações próximas da média, esses fatores não foram suficientes para posicioná-la de forma competitiva no conjunto geral.

Recomendação: Vender a Loja 4.
Essa ação permitirá concentrar recursos nas unidades mais lucrativas e com melhor reputação, além de viabilizar o investimento no novo empreendimento com maior potencial de retorno.
